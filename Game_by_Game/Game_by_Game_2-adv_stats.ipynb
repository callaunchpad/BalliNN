# Calculations

```python
""" gets the season games given the seasonID and abbreviation of the team """
""" Season is based on the year that the season STARTED in. I.E 2016-2017 has seasonID 2016"""
""" Reg season starts with 2, playoffs 4, pre 1 """
""" 
For some reason summer league and preseason games are sometimes included in reg season IDs.
The function accounts for this by only looking at Game IDs that start with 002 for reg reason,
004 for playoffs, 001 for preseason
"""

def get_season_team(season, team_id, team_names):
    gameID_prefix = 2
    return team_names[team_id].loc[(team_names[team_id]["season"].str.contains(season)) & \
                                    (team_names[team_id]['GAME_ID'].str.startswith('00{0}'.format(gameID_prefix)))]


""" gets the number of rows in the passed df"""
def num_rows(df):
    return len(df.index)

""" returns the number of features of the df"""
def num_features(df):
    return len(df.columns)

""" returns a list of the features from the start index moving forward. Using 9 for now """
def get_features(df, start_idx=6):
    return [feature for feature in df.columns.tolist()[start_idx:]]

""" 
util func: gets last n games given the curr game
in the df with the correct season, year
"""
def last_n_games(df, n, curr_game):
    n_rows = num_rows(df)
    # the first row is the latest game. I.E for reg season it's 82nd game

def not_first_game(games, game_id):
    # get the row index of the game_id. Should already be reindexed from find_diff_vector
    game_idx = games[games['GAME_ID'] == game_id].index.values.astype(int)[0]
    return game_idx < num_rows(games) - 1
    
"""
returns avg team stats vector over the last n games given
the current game. Assumes games 0 indexed. Takes in a df games of all the games a team played.
"""
def get_team_avgs(games, game_id, n):
    start_idx = 6
    # get the row index of the game_id. Should already be reindex from find_diff_vec so its 0-num_rows-1
    game_idx = games[games['GAME_ID'] == game_id].index.values.astype(int)[0]
    
    # get the past n games using the game_idx. df is sorted with the latest games 1st, which is 
    # why you start at game_idx + 1 (don't include curr game in avg obviously)
    past_games = games.iloc[game_idx + 1:min(game_idx + n + 1, num_rows(games))]
    #convert df to np array
    past_games = np.array(past_games)
    # drop the first 6 entries for each row
    past_games = past_games[:, start_idx:]
    return np.nanmean(past_games, axis=0)

def print_team_avgs(season_games, team_avgs):
    features = get_features(season_games)
    for stat, feature in zip(team_avgs, features):
        print("{0}:{1}".format(feature, stat))
```

```python
"""
return the difference vector betweeen the current team and its opponent
If it is the first game for either team, return Nones
"""
# game_id: the id for the game you wanna find the diff vector for
# n: the number of previous games to use for stat avgs
# season_idx: the idx of the season in the seasons list that this game was played in. Never can be 0
def find_diff_vector(game_id, n, season_idx):
    diff, outcome = None, None
    
    # get the two teams that played in the game using game_id
    team_rows = new_adv_df.loc[lambda df: df["GAME_ID"] == game_id]
    team_id_1 = team_rows.iloc[0]["TEAM_ID"]
    team_id_2 = team_rows.iloc[1]["TEAM_ID"]
    
    team1_games = team_names_test[team_id_1]
    team2_games = team_names_test[team_id_2]
    team1_games.reset_index(drop=True, inplace=True)
    team2_games.reset_index(drop=True, inplace=True)
    
    # check that this isn't the first game for either team
    if (not_first_game(team1_games, game_id) and not_first_game(team2_games, game_id)):
        team1_avgs = get_team_avgs(team1_games, game_id, n)
        team2_avgs = get_team_avgs(team2_games, game_id, n)

        team1_pts = team_rows.iloc[0]["PTS"]
        team2_pts = team_rows.iloc[1]["PTS"]

        # determine outcome based which team scored more points
        outcome = 1 if team1_pts > team2_pts else 0

        # normalize 
        team1_norm = normalize(seasons[season_idx - 1], team1_avgs)
        team2_norm = normalize(seasons[season_idx - 1], team2_avgs)

        diff = np.subtract(team1_norm, team2_norm)

    return diff, outcome
```

```python
""" gets diff vectors for all the games 2-82 for given team and season""" 
# for each season, iterate through all game ids that start with 002xy
# where x y correspond to the season. Feed the outputs into the model
adv_df_ind_reset = new_adv_df.reset_index(drop=True)
def get_season_diff_vectors(n):
    diff_vectors = []
    labels = []
    for i, season in enumerate(seasons[1:]):
        print('processing season: {0}'.format(season))
        start_year = season[2:4]
        for j, row in adv_df_ind_reset.iterrows():
            game_id = row["GAME_ID"]
            if (game_id[3:5] == start_year):
#                 print(game_id)
                diff, outcome = find_diff_vector(game_id, n, i)
                if (diff is not None and outcome is not None):
                    diff_vectors.append(diff)
                    labels.append(outcome)
        print('finished season: {0}'.format(season))
    return diff_vectors, labels
```
